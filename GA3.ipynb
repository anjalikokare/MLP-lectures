```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.feature_selection import VarianceThreshold
from sklearn.pipeline import FeatureUnion
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.pipeline import FeatureUnion
```

```python
data = pd.read_csv('Week3_GA_dataset.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data = data.replace({'?':np.nan})

data['V1'] = data['V1'].astype('float64')
data['V2'] = data['V2'].astype('float64')
```

```python
data.info()
```

```python
impute_df = [('Simple Imputer',SimpleImputer(strategy='mean',missing_values=np.nan),[0,1])]

impute_df = ColumnTransformer(impute_df,remainder='passthrough',verbose_feature_names_out=False).set_output(transform='pandas')


scale_df = [('Standard Scaler',StandardScaler(),[0,1,2,3])]

scale_df = ColumnTransformer(scale_df,remainder='drop',verbose_feature_names_out=False).set_output(transform='pandas')

pipe1 = Pipeline(
    [('Imputation',impute_df),
    ('scalling',scale_df)]).set_output(transform='pandas')

pipe1
```

```python
encode_df = [('Ordinal Encoder',OrdinalEncoder(),[4])]

encode_df = ColumnTransformer(encode_df,remainder='drop',verbose_feature_names_out=False).set_output(transform=False)

pipe2 = Pipeline([
    ('Encoder',encode_df)
])
pipe2
```

```python
union_pipe = FeatureUnion([
    ('Imputing and Scalling Pipe',pipe1),
    ('Encoding Pipe',pipe2)
]).set_output(transform='pandas')

union_pipe
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
full_pipe = Pipeline([('Preprocessing',union_pipe),
                     ('VarienceThreshold',VarianceThreshold(threshold=0.1))]).set_output(transform='pandas')

full_pipe
```

```python
X = data.drop('Target',axis=1)
y = data['Target']
```

```python
full_pipe.fit_transform(X).shape
```

```python
X_changed = full_pipe.fit_transform(X)
X_changed.head(2)
```

```python
X['V5'].value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
```

```python
new_X = union_pipe.fit_transform(X)
new_X
```

```python
oe = OrdinalEncoder()
y_encoded = oe.fit_transform(data[['Target']])

rfe = RFE(LogisticRegression(),n_features_to_select = 2)
rfe.fit_transform(new_X,y_encoded.reshape(-1))
```

```python
rfe.support_
```

```python
from sklearn.feature_selection import SequentialFeatureSelector as SFS

sfs = SFS(LogisticRegression(),n_features_to_select = 2,direction='forward')
sfs.fit_transform(new_X,y_encoded.reshape(-1))
```

```python
sfs.support_
```

```python
sfs = SFS(LogisticRegression(),n_features_to_select = 2,direction='backward')
sfs.fit_transform(new_X,y_encoded.reshape(-1))
```
